## 改行をなくして，文字列を数値型に変換する．

```python
input_str = input_str.replace('\n', '')
input_arr = np.array([int(i) for i in list(input_str)], dtype=np.float64)
print(input_arr)
```

## 初期値設定

```python
greatest_product = 0
adjacent_digits = None
```

## 最大値を計算する

```python
for i in range(0, input_arr.shape[0] - 13):
    arr = input_arr[i:i + 13]
    product = reduce(operator.mul, arr)
#         print(product)
    if product > greatest_product:
        greatest_product = product
        adjacent_digits = arr

ans = greatest_product
```

## 方針

1000までの自然数を生成  
3または5の倍数を抽出  
抽出した数の合計を算出  

## インポート

```python
import numpy as np
```

## 1000までの自然数を生成

```python
input_arr = np.arange(1, 1000)
print(input_arr)
```

## input_valから3の倍数または5の倍数にあたる数を抽出し，ext_valに格納する

```python
ext_arr = input_arr[(input_arr%3 == 0) | (input_arr%5 == 0)]
print(ext_arr)
```

## ext_val に格納されている数の合計を算出する

```python
ans = np.sum(ext_arr)
```

## 答え

```python
print(ans)
```
